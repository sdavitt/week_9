#### Select all rows with records of slashed values

```python
for i in range(len(df_18[df_18['Fuel'].str.contains('/')])):
    flist = df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Fuel'].split('/')
    cmlist = df_18[df_18['Fuel'].str.contains('/')].iloc[i]['City_MPG'].split('/')
    hmlist = df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Hwy_MPG'].split('/')
    blist = df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Cmb_MPG'].split('/')
    #print(f'{flist}{cmlist}{hmlist}{blist}')
    df_18 = df_18.append({
        'Model': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Model'], 
        'Displ': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Displ'], 
        'Cyl': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Cyl'], 
        'Trans': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Trans'], 
        'Drive': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Drive'], 
        'Fuel': flist[1], 
        'Stnd': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Stnd'], 
        'Underhood_ID': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Underhood_ID'], 
        'Veh_Class': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Veh_Class'], 
        'Air_Pollution_Score': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Air_Pollution_Score'],
        'City_MPG': cmlist[1],
        'Hwy_MPG': hmlist[1], 
        'Cmb_MPG': blist[1],
        'Greenhouse_Gas_Score': df_18[df_18['Fuel'].str.contains('/')].iloc[i]['Greenhouse_Gas_Score']
    }, ignore_index=True)
df_18['Fuel'] = df_18['Fuel'].apply(lambda x: x.split('/')[0])
df_18['City_MPG'] = df_18['City_MPG'].apply(lambda x: x.split('/')[0])
df_18['Hwy_MPG'] = df_18['Hwy_MPG'].apply(lambda x: x.split('/')[0])
df_18['Cmb_MPG'] = df_18['Cmb_MPG'].apply(lambda x: x.split('/')[0])
```
